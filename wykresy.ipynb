```python
import matplotlib.pyplot as plt

# Dane
resolutions = ["16x16", "32x32", "64x64", "128x128", "256x256", "512x512"]
accuracy = [0.9781, 0.9872, 0.9854, 0.9927, 0.9936, 0.9945]
precision = [0.9810, 0.9900, 0.9887, 0.9941, 0.9949, 0.9956]
recall = [0.9597, 0.9710, 0.9654, 0.9817, 0.9845, 0.9873]
f1_score = [0.9693, 0.9799, 0.9760, 0.9876, 0.9895, 0.9913]

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.plot(resolutions, accuracy, marker='o', label='Accuracy')
plt.plot(resolutions, precision, marker='o', label='Precision')
plt.plot(resolutions, recall, marker='o', label='Recall')
plt.plot(resolutions, f1_score, marker='o', label='F1 Score')

plt.title('Metryki klasyfikacji względem rozdzielczości obrazu')
plt.xlabel('Rozdzielczość obrazu')
plt.ylabel('Wartość metryki')
plt.ylim(0.95, 1.0)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
